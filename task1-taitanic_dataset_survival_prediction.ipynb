```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

taitanic_data['Sex'] = labelencoder.fit_transform(taitanic_data['Sex'])
taitanic_data['Age'] = labelencoder.fit_transform(taitanic_data['Age'])

taitanic_data.head(10)
```

```python
taitanic_data['Sex'], taitanic_data['Survived'], taitanic_data['Age']
```

```python
x=taitanic_data[['Pclass', 'Sex', 'Age']]
y=taitanic_data['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression(random_state=0)
log.fit(x_train, y_train)
```

```python
pred=print(log.predict(x_test))
```
